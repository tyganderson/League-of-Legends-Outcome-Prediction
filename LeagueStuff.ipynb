```python
import requests
import pandas as pd
```

```python
response = requests.get('http://ddragon.leagueoflegends.com/cdn/13.7.1/data/en_US/champion.json')
display(response)
```

```python
display(response.json())
```

```python
json = response.json()
# display(json['data'])
champs = json['data']
table = []
for champ in champs:
    table.append([int(champs[champ]['key']),champs[champ]['id']])
#     print(champs[champ]['key'], ':', champs[champ]['id'])
    
display(table)
```

```python
def myFunc(e):
  return e[0]

table.sort(key=myFunc)
```
